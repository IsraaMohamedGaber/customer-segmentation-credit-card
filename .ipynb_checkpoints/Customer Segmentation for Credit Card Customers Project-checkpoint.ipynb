# Phase 2: Determining the Optimal Number of Clusters


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Elbow Method to find the optimal number of clusters for K-Means
wcss = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)  # Sum of squared distances
# elbow
plt.figure(figsize=(8, 5))
plt.plot(K, wcss, 'bo-', linewidth=2)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
#WCSS starts decreasing more slowly
```

```python
#Silhouette Score
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    score = silhouette_score(df_scaled, kmeans.labels_)
    print(f"Silhouette Score for k={k}: {score:.4f}")
    
#Score range: -1 to 1
#Closer to 1 = well-clustered
#Around 0 = overlapping clusters
#Negative = wrong clustering (data might be in the wrong cluster)
```

# phase 3: Customer Segmentation

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Use optimal k from Silhouette Score based on pervious step
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
cluster_labels = kmeans.fit_predict(df_scaled)
# Add cluster labels to the original DataFrame
df['Cluster'] = cluster_labels
```

```python
cluster_summary = df.groupby('Cluster').mean().round(2)
cluster_summary
```

```python
df['Cluster'].value_counts()
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]
# Scatter plot of clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set1')
plt.title('Customer Segments Visualized with PCA')
plt.show()
```

```python
# Normalize the data for radar chart
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(cluster_summary)

labels = cluster_summary.columns
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop

# Plot each cluster
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for i in range(k):
    values = normalized_data[i].tolist()
    values += values[:1]
    ax.plot(angles, values, label=f'Cluster {i}')
    ax.fill(angles, values, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)
ax.set_title('Radar Chart of Cluster Profiles', size=15)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```

# Phase 5: Business Insights & Strategic Recommendations

#  Cluster 0: Low Usage Customers
- **Profile**: Low balance, low credit usage, minimal transactions
- **Insight**: These customers show little engagement or usage.
- **Strategy**: Run welcome/onboarding campaigns with usage incentives.

###  Cluster 1: Consistent, Moderate Spenders
- **Profile**: Moderate balance, steady purchases, regular payments
- **Insight**: Reliable and stable customers.
- **Strategy**: Encourage loyalty through cashback or referral bonuses.

###  Cluster 2: High Spenders / High Variability
- **Profile**: High purchase volume or high cash advances, diverse usage
- **Insight**: High revenue but potentially higher risk.
- **Strategy**: Offer premium rewards to responsible payers, monitor credit risk for others.

---

# Business Insights
- Customer Segmentation: The clustering analysis has revealed distinct customer segments based on their purchasing behavior, credit usage, and payment patterns. Understanding these segments allows for targeted marketing strategies.

- High-Value Customers: Some clusters exhibit high purchasing power and frequent transactions. These customers are likely the most profitable and should be prioritized for loyalty programs and premium services.

- Low-Engagement Customers: Other clusters show minimal activity, such as low purchase frequency and high cash advances. These customers may require re-engagement strategies or cost-saving measures.

- Payment Behavior: Clusters with high balances but low payment rates may indicate financial stress. Offering flexible payment plans or financial advice could improve their engagement and reduce defaults.

- Credit Utilization: Customers who frequently max out their credit limits may be at higher risk of default. Monitoring and adjusting their credit limits could mitigate risks.

---

### Strategic Recommendations
- Personalized Marketing Campaigns:

High-Value Customers: Offer exclusive rewards, discounts, or premium services to retain their loyalty.

Low-Engagement Customers: Use targeted promotions or cashback incentives to encourage more frequent purchases.

- Credit Limit Adjustments:

For customers with high credit utilization, consider offering temporary credit limit increases or financial planning tools to help them manage their spending.

- Financial Education:

Provide resources or workshops on responsible credit usage, especially for clusters with high cash advances or low payment rates.

- Dynamic Pricing and Offers:

Tailor pricing and offers based on cluster behavior. For example, offer installment plans to customers who prefer spreading out payments.

- Risk Management:

Implement early warning systems for customers showing signs of financial distress, such as missed payments or high balances.

- Product Recommendations:

Use clustering insights to recommend products or services that align with the purchasing habits of each segment.


- Use customer segmentation in CRM to tailor marketing
- Focus retention strategies on high-value customers (Cluster 2)
- Design risk-mitigation tools for high cash advance users
